```python

import re

# Assuming you already have the DataFrame df_st

def extract_last_number(text):
    # Regular expression to find numbers, allowing 'K' for thousands
    matches = re.findall(r'(\d+\.?\d*)K?', text)

    if matches:
        # Return the last match, converting it to a float (and handling 'K')
        last_match = matches[-1]
        if 'K' in text:
            return float(last_match)  # Return as a number, already in thousand
        else:
            return float(last_match)
    return None

# Apply the function to the 'Box Finish' column
df_st_o['Box Finish'] = df_st_o['Box Finish'].apply(extract_last_number)
```

```python
df_st_o['Box Finish'] = df_st_o['Box Finish'] * 1000
```

```python
df_st_o['Issue'] = pd.to_datetime(df_st_o['Issue'], format='%B %Y').dt.year
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df_st_o['Cover'] = label_encoder.fit_transform(df_st_o['Cover'])
cover_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print(cover_mapping)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming 'df_st_o' is your DataFrame

# Features to plot against the target
features = ['Cover', 'RG', 'Diff', 'Int']

# Setting up the plotting area
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
fig.tight_layout(pad=5.0)

# Creating scatter plot for each feature
for i, feature in enumerate(features):
    ax = axes[i//3, i%3]  # Determine row and column position on the grid
    sns.scatterplot(data=df_st_o, x=feature, y='St-O', ax=ax)
    ax.set_title(f'Scatter Plot: {feature} vs St-O')
    ax.set_xlabel(feature)
    ax.set_ylabel('St-O')

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Selecting the features and targets
X = df_st_o[['Cover', 'RG', 'Diff', 'Int']]  # Features
y = df_st_o['St-O']  # Targets

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Displaying shapes of the resulting datasets
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Create an instance of GradientBoostingRegressor
# Here, you can tweak the hyperparameters like n_estimators, learning_rate, and max_depth
gbr = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, max_depth=2, random_state=42)

# Fit the model on the training data
gbr.fit(X_train, y_train)

# Predict the St-O values on the test set
y_pred = gbr.predict(X_test)

# Calculate the mean squared error and R^2 score to evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Output the performance metrics
print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Create an instance of RandomForestRegressor
# Parameters like n_estimators, max_depth, and random_state can be tuned
rf_model = RandomForestRegressor(n_estimators=300, max_depth=5, random_state=42)

# Fit the RandomForest model on the training data
rf_model.fit(X_train, y_train)

# Predict the St-O values on the test set
y_pred_rf = rf_model.predict(X_test)

# Calculate the mean squared error and R^2 score to evaluate the model
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

# Output the performance metrics
print(f'Mean Squared Error (Random Forest): {mse_rf}')
print(f'R^2 Score (Random Forest): {r2_rf}')
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

# Create an instance of SVR
# The 'rbf' kernel is commonly used for nonlinear data, 'C' and 'epsilon' can be tuned
svr_model = SVR(kernel='rbf', C=3, epsilon=0.01)

# Fit the SVR model on the training data
svr_model.fit(X_train, y_train)

# Predict the St-O values on the test set
y_pred_svr = svr_model.predict(X_test)

# Calculate the mean squared error and R^2 score to evaluate the model
mse_svr = mean_squared_error(y_test, y_pred_svr)
r2_svr = r2_score(y_test, y_pred_svr)

# Output the performance metrics
print(f'Mean Squared Error (SVR): {mse_svr}')
print(f'R^2 Score (SVR): {r2_svr}')
```

```python
from sklearn.ensemble import StackingRegressor, GradientBoostingRegressor, RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR

# Define base models
base_models = [
    ('svr', SVR(C=3.0, epsilon=0.01)),
    ('gbr', GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, max_depth=2, random_state=42))
]

# Define the meta-model
meta_model = LinearRegression()

# Create the stacking model
stacked_model = StackingRegressor(estimators=base_models, final_estimator=meta_model, cv=5)

# Fit the stacked model
stacked_model.fit(X_train, y_train)

# Evaluate the model
y_pred_stack = stacked_model.predict(X_test)
mse_stack = mean_squared_error(y_test, y_pred_stack)
r2_stack = r2_score(y_test, y_pred_stack)

print(f'Mean Squared Error (Stacking): {mse_stack}')
print(f'R^2 Score (Stacking): {r2_stack}')

```
